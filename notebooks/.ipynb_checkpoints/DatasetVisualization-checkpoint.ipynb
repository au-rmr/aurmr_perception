```python
from datasets import AURMRDataset, collate, denormalize_rgb
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
```

```python
train_data = AURMRDataset('../data/5k_dataset/train_shard_000000.h5', patch_size=32, max_nobj=10)
```

```python
batch_size=10
n_worker=1
batch_idx = 1

train_loader = DataLoader(
    train_data,
    batch_size,
    collate_fn=collate,
    num_workers=n_worker,
    pin_memory=True
)
```

```python
i, data = next(enumerate(train_loader))
```

```python
data['rgb0'].shape
```

```python
f, axarr = plt.subplots(1,2)
axarr[0].imshow(denormalize_rgb(data['rgb0'][batch_idx]))
axarr[1].imshow(denormalize_rgb(data['rgb1'][batch_idx]))
```

```python
f, axarr = plt.subplots(1,10)
for i in range(10):
    obj_mask = data['input_masks'][batch_idx][i]
    axarr[i].imshow(obj_mask)
```

```python
f, axarr = plt.subplots(1,10)
for i in range(10):
    obj_mask = data['label_masks'][batch_idx][i]
    axarr[i].imshow(obj_mask)
```

```python
f, axarr = plt.subplots(1,2)
axarr[0].imshow(data['depth0'][batch_idx][50:200,50:200])
axarr[1].imshow(data['depth1'][batch_idx][50:200,50:200])
```
